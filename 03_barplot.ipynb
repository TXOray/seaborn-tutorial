```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mart=pd.read_excel('/content/drive/MyDrive/supermarket_sales.xlsx')
mart.columns
#mart.head()
```

```python
sns.get_dataset_names()#get example datas
titanic=sns.load_dataset('titanic')
#titanic
```

```python
sns.barplot(y='Product line',x='Total',data=mart) #get the average sells  dark line is error bar shows variation of data
```

```python
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',data=mart,hue='Gender')#add hue to the bar
```

```python
#plot the bars in a given order
plt.figure(figsize=(15,5))


sns.barplot(x='Product line',y='Total',hue='Gender',hue_order=['Male','Female'],data=mart,order=sorted(mart['Product line'].unique()))#sorted returns a list
```

```python
#add cap on error bar
plt.figure(figsize=(15,5))

sns.barplot(x='Product line', y='Total',hue='Gender',hue_order=['Male','Female'],data=mart,order=sorted(mart['Product line'].unique()),capsize=0.2)
```

```python
#remove confidence interb=val

plt.figure(figsize=(15,5))


sns.barplot(x='Product line',y='Total',hue='Gender',hue_order=['Male','Female'],data=mart,order=sorted(mart['Product line'].unique()),capsize=0.2,ci=None)
```

```python
#change colors
plt.figure(figsize=(15,5))


sns.barplot(x='Product line',y='Total',hue='Gender',hue_order=['Male','Female'],data=mart,order=sorted(mart['Product line'].unique()),capsize=0.2,ci=None,color='red')
```

```python
#change color using pallete attribute
plt.figure(figsize=(15,5))


sns.barplot(x='Product line',y='Total',data=mart,hue='Gender',order=sorted(mart['Product line'].unique()),capsize=0.2,ci=None,palette='cividis')
```

```python
#Using saturation
plt.figure(figsize=(15,5))


sns.barplot(x='Product line',y='Total',data=mart,hue='Gender',order=sorted(mart['Product line'].unique()),capsize=0.2,ci=None,palette='cividis',saturation=0.5)
```

```python
#Change default aggregation method
plt.figure(figsize=(15,5))


sns.barplot(x='Product line',y='Total',data=mart,hue='Gender',order=sorted(mart['Product line'].unique()),capsize=0.2,ci=None,palette='cividis',estimator=np.median)#estimator
```
