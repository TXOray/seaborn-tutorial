```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mart=pd.read_excel('/content/drive/MyDrive/supermarket_sales.xlsx')
mart.columns
#mart.head()
```

```python
#Create a basic COUNTPLOT to get number of transaction for each of the product line
#and change the figure size
plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart)
```

```python
#make it horizontal barplot
sns.countplot(y='Product line',data=mart)
```

```python
#Add hue to get the count on two categories Gender and Product line
plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart,hue='Gender')
```

```python
#using diffrent palette

plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart,hue='Gender',palette='GnBu')
```

```python
#.Change Style using facecolor, linewidth and edge color
plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart
              ,facecolor=(1,0,1,0)#Red Green Blue Alph=transprancy 強度都是0~1 0 完全透明
              ,linewidth=5
              ,edgecolor=sns.color_palette('dark',10))
```
